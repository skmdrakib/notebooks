```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Simulate quality control data (defects per batch)
np.random.seed(42)
batches = np.random.poisson(lam=2, size=1000)  # Poisson distribution to simulate defect counts
```

**Task1:** Create quality control data of defects per batch with poisson distribution with lam=5 and repeat the steps

```python
# Convert to DataFrame
data = pd.DataFrame(batches, columns=['defects'])
data
```

```python
# Number of samples and sample size
num_samples = 500
sample_size = 30
```

**Task2:** take sample size 50 and sample size 10 and repeat the steps

```python
# Store sample means
sample_means = []
```

```python
# Drawing samples and calculating the means
for _ in range(num_samples):
    sample = np.random.choice(batches, sample_size, replace=False)
    sample_means.append(np.mean(sample))
```

```python
# Plotting the distribution of original data and sample means
plt.figure(figsize=(14, 6))

# Original data distribution
plt.subplot(1, 2, 1)
sns.histplot(data['defects'], bins=20, kde=True, color='lightblue')
plt.title('Distribution of Defects per Batch')
plt.xlabel('Number of Defects')
plt.ylabel('Frequency')
```

```python
# Sample means distribution
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=20, kde=True, color='salmon')
plt.title('Distribution of Sample Means')
plt.xlabel('Mean Defects per Batch')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
