```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Generate sample data (weights of parts)
np.random.seed(42)
sample_weights = np.random.normal(loc=51, scale=2, size=30)  # Mean 51, std dev 2, sample size 30
sample_weights
```

**Task:** create a dataset with mean 45 and std 4 and follow the steps

```python
# Define the target mean
target_mean = 50
```

```python
# Perform the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_weights, target_mean)
p_value
```

```python
# Set significance level
alpha = 0.05
```

```python
# Conclusion based on p-value
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
    print("There is evidence that the mean weight of parts is different from 50 grams.")
else:
    print("Fail to reject the null hypothesis (H0).")
    print("There is no evidence that the mean weight of parts is different from 50 grams.")
```

```python
# Visualize the sample data
plt.figure(figsize=(10, 6))
plt.hist(sample_weights, bins=10, color='lightblue', edgecolor='black')
plt.axvline(target_mean, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of Part Weights')
plt.xlabel('Weight (grams)')
plt.ylabel('Frequency')
plt.show()
```
