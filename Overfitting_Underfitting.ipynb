```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.2, X.shape[0])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Linear regression model
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

# Predictions
y_train_pred = linear_regressor.predict(X_train)
y_test_pred = linear_regressor.predict(X_test)

# Plotting
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_train_pred, color='red', label='Linear fit')
plt.legend()
plt.title('Linear Regression (Underfitting)')
plt.show()

# Performance
print("Linear Regression MSE (Train):", mean_squared_error(y_train, y_train_pred))
print("Linear Regression MSE (Test):", mean_squared_error(y_test, y_test_pred))
```

```python
# Polynomial regression model with a high degree
poly_features = PolynomialFeatures(degree=20)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.transform(X_test)

poly_regressor = LinearRegression()
poly_regressor.fit(X_poly_train, y_train)

# Predictions
y_train_pred_poly = poly_regressor.predict(X_poly_train)
y_test_pred_poly = poly_regressor.predict(X_poly_test)

# Plotting
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_train_pred_poly, color='red', label='Polynomial fit')
plt.legend()
plt.title('Polynomial Regression (Overfitting)')
plt.show()

# Performance
print("Polynomial Regression MSE (Train):", mean_squared_error(y_train, y_train_pred_poly))
print("Polynomial Regression MSE (Test):", mean_squared_error(y_test, y_test_pred_poly))
```
